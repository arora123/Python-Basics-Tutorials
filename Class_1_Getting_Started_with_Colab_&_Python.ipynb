# Key board shortcuts

1. to run current cell - Cmd/Ctrl+Enter

2. to run current cell & move to next cell - Shift+Enter

3. to run current cell & insert a new code cell below it - Alt+Enter [both works same]

4. comment/uncomment single line -#

5. comment/uncomment multiple lines - Ctrl+/

6. Sometimes, you accidentally delete a cell (cut the cell) , now what to do ? Is there any way you can get it back(whole code and output) ?

Use ctrl + shift +z

7. find & replace - Ctrl + H

8. To try a code which you don't want to be a part of your notebook - Ctrl + Alt + N or Insert - scratch code cell

9. Show/Hide code [Edit menu] - Alt + L Or use hash@title 

10. Edit - Clear all output

Edit - change runtime type - GPU/TPU
